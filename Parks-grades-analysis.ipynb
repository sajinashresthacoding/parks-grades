```python

##let's create a filter to check to see if there are any data that holds the 2022 year
df_grades['Inspection Year'].eq(2022)
```

```python
##now let's put the filter inside a dataframe so that we get a dataframe that is only the 2022 inspections

df_22 = df_grades[df_grades['Inspection Year'].eq(2022)]
df_22
```

```python
##now that we have this dataframe of just 2022, let's create another filter that filters all of 2022 inspections
## only for queens
queens_filter = df_22["Prop ID"].str.startswith("Q")

##now lets put that filter inside the df_22 and create a new dataframe that filters all of 2022 inspections
## only for queens
df_q22 = df_22[queens_filter].copy()
df_q22
```

```python
##now let's call the maps to combine the two dataframes to see the area of these parks
df_mp = pd.read_csv("data/PIP-MAPPED.csv")
df_mp
```

```python
##woah! that's a lot!
##we only need prop ID, zipcode, and category tho. so let's only pull that from df_mp
df_mp[["Prop ID" , "ZipCode" , "Category"]]
```

```python
##now let's merge this altered df_mp with our df_q22 with prop id as the connecting factor
## this way we can combine and see both the grades and zipcodes of these parks

df_q22_zp = df_q22.merge(df_mp[["Prop ID" , "ZipCode" , "Category"]], on = "Prop ID").copy()
df_q22_zp
```

```python
##but we are only investigating the grades of certain zipcodes in queens
## let's get those zipcodes by creating a filter
##begin with a list

q_zipcodes = ["11372" , "11373" , "11377"]
q_zipcodes
```

```python
##let's create a filter! 
zp_filter = df_q22_zp["ZipCode"].isin(q_zipcodes)
##now let's put that filter in our dataframe and create a new dataframe with the filter
df_ta = df_q22_zp[zp_filter]
df_ta
```

```python
##now let's ask the new dataframe (df_ta) to show us the most recent inspections.
##because these inspections happen every year multiple times, let's only go with the msot recent inspection

df_mri = df_ta.sort_values(by = 'Date') \
.drop_duplicates(subset = 'Prop ID' , keep = 'last') 
## ^ a little confused here. what are we doing with keep = last? 
df_mri
```

```python
##now that we have what we need, let's see how the grades are like 
df_mri[['Category','Overall Condition']].value_counts().unstack()
```
