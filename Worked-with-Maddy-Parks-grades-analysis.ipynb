```python
df_grades['Inspection Year']
```

```python
##this is just creating the filter
df_grades['Inspection Year'].eq(2022)
```

```python
##this basically takes ^ trues and falses and filters and keeps only stuff that is true
##this is putting the filter inside the dataframe
df_22 = df_grades[df_grades['Inspection Year'].eq(2022)].copy()
df_22
```

```python
##all of queens' inspections in 2022
queens_filter = df_22["Prop ID"].str.startswith("Q")
df_q22 = df_22[queens_filter].copy()
df_q22
```

```python
##always do this to double check  you actually has the highest datetime because pd
##doesnt understand the concept of date and time
df_grades['Date'] = df_grades['Date'].apply(pd.to_datetime)
```

```python
df_grades['Date'].max()
```

```python
df_mp = pd.read_csv("data/PIP-MAPPED.csv")
df_mp
```

```python
##reducing the amount to only the columns we need in mp
df_mp[["Prop ID" , "ZipCode" , "Category"]]
```

```python
##combined data
df_q22_zp = df_q22.merge(df_mp[["Prop ID" , "ZipCode" , "Category"]], on = "Prop ID").copy()
df_q22_zp
```

```python
q_zipcodes = ["11372", "11373", "11377"]
q_zipcodes
```

```python
## [] is for one column and [[]] is more than 1 column
zp_filter = df_q22_zp["ZipCode"].isin(q_zipcodes)
zp_filter
```

```python
df_ta = df_q22_zp[zp_filter]
df_ta
```

```python
##shows us most recent inspection 
df_mri = df_ta.sort_values(by = 'Date') \
.drop_duplicates(subset = 'Prop ID',keep = 'last')

df_mri
```
